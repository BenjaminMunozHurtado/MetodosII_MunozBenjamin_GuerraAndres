```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(u, y, q):
        return u ** q
```

```python
def GetRungeKutta(f, x, y0, q):
    h = x[1] - x[0]
    y = np.zeros_like(x)
    y[0] = y0
    
    for i in range(1, len(x)):
        k1 = h * f(x[i-1], y[i-1], q)
        k2 = h * f(x[i-1] + 0.5 * h, y[i-1] + 0.5 * k1, q)
        k3 = h * f(x[i-1] + 0.5 * h, y[i-1] + 0.5 * k2, q)
        k4 = h * f(x[i-1] + h, y[i-1] + k3, q)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        
    return y
```

```python
lista_q = [0., 0.2, 0.4, 0.7, 0.9, 1.]
N = 1000
t = np.linspace(0, 10, N)
```

```python
lista_ys = []

for q in lista_q:
    yrk = GetRungeKutta(f, t, 0, q)
    lista_ys.append(yrk)
```

```python
for i, y in enumerate(lista_ys):
    plt.plot(t, y, label=f"q={lista_q[i]}")

plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```
